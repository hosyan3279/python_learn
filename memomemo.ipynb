クラス定義

```python
class Greeting:
    def __init__(self, name):#自分に名前を入れている
        self.name = name

    def say_hello(self):#say hello関数
        print("hello " + self.name)

paiza = Greeting("paiza")#クラス名paiza
paiza.say_hello()
python = Greeting("python")#クラス名python
python.say_hello()
```
